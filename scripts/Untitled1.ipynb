```python
# Canais que ser√£o usados
use = ['EEG FP1-REF', 'EEG FP2-REF', 'EEG F3-REF', 'EEG F4-REF', 'EEG C3-REF', 'EEG C4-REF',
       'EEG P3-REF', 'EEG P4-REF', 'EEG O1-REF', 'EEG O2-REF', 'EEG F7-REF', 'EEG F8-REF', 
       'EEG T3-REF', 'EEG T4-REF', 'EEG T5-REF', 'EEG T6-REF', 'EEG T1-REF', 'EEG T2-REF',
       'EEG FZ-REF', 'EEG CZ-REF', 'EEG PZ-REF', 'EEG C3P-REF', 'EEG C4P-REF', 'EEG SP1-REF',
       'EEG SP2-REF', 'EEG 29-REF', 'EEG 30-REF', 'EEG 31-REF', 'EEG 32-REF']
```

### Concatenando janelas de EEG com crise

Estabelecer janelas de 0.25 segundos

```python
import mne

path = '/home/davi/Transfer√™ncias/www.isip.piconepress.com/projects/tuh_eeg/downloads/tuh_eeg_seizure/v1.5.0/edf/'

fs = int(np.unique(w['fs']))
w_windows = np.zeros((len(use),0))


for i in range(len(w)):
    raw = mne.io.read_raw_edf(path+w.iloc[i,0].replace('tse','edf'),preload=False, verbose=False)
    d = raw.to_data_frame().T
    windows = d.loc[use,int(fs*.25)*int(w.iloc[i,1]):int(fs*.25)*int(w.iloc[i,2])].to_numpy()
    w_windows = np.hstack((w_windows,windows))
    print(w_windows.shape)
```

```python
n_windows = np.zeros((len(use),1))

for i in range(len(n)):
    raw = mne.io.read_raw_edf(path+n.iloc[i,0].replace('tse','edf'),preload=False, verbose=False)
    d = raw.to_data_frame().T
    windows = d.loc[use,:].to_numpy()
    n_windows = np.hstack((n_windows,windows))
    print(n_windows.shape)
    if n_windows.shape[1]>=w_windows.shape[1]:
        break
```

```python
print(w_windows.shape[1])
print(n_windows.shape[1])
```

```python
w_windows = w_windows[:,1:]
n_windows = n_windows[:,1:]
```

```python
n_windows = n_windows[:,:w_windows.shape[1]]
print(w_windows.shape==n_windows.shape)
```

```python
np.save('w_windows.npy',w_windows)
np.save('n_windows.npy',n_windows)
```
