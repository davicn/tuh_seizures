```python
import numpy as np
import pandas as pd
import os

path  = os.getcwd()
path2 = '/home/pib3/Projetos/tuh_eeg_seizures/dev_test/01_tcp_ar/'


file = np.load(path + '/docs/data_ecg_emg.npy')
col = ['paths','freq','montage','type','start','end','class']
data = pd.DataFrame(data=file,columns=col)
print(data.shape)
data 
```

```python
np.unique(data['montage'][0])
```

```python
use = ['EEG C3-REF', 'EEG C4-REF','EEG CZ-REF', 'EEG EKG1-REF', 
       'EEG F3-REF', 'EEG F4-REF','EEG F7-REF', 'EEG F8-REF', 
       'EEG FP1-REF', 'EEG FP2-REF','EEG FZ-REF', 'EEG O1-REF', 
       'EEG O2-REF', 'EEG P3-REF','EEG P4-REF', 'EEG PZ-REF', 
       'EEG T1-REF', 'EEG T2-REF', 'EEG T3-REF', 'EEG T4-REF',
       'EEG T5-REF', 'EEG T6-REF', 'EMG-REF']
```

```python
com_crise = data[data['type']!='bckg']
sem_crise = data[data['type']=='bckg']

com_crise.index = np.arange(len(com_crise))
sem_crise.index = np.arange(len(sem_crise))

print(com_crise.shape)
print(sem_crise.shape)
```

### Concatenando janelas com crise

```python
com_crise.head(3)
```

```python
inicio = []
fim = []

for i in range(len(com_crise)):
    aux_i = []
    aux_f = []
    for ii in range(1,len(com_crise.loc[i,'start'])):
        if ii%2!=0:
            aux_i.append(com_crise.loc[i,'start'][ii])
            aux_f.append(com_crise.loc[i,'end'][ii])
    inicio.append(np.array(aux_i))
    fim.append(np.array(aux_f))
```

```python
for i in range(len(inicio)):
    print(inicio[i],fim[i])
    print()
```

```python
import mne

fs = 256
mx_w = np.ones((23,1))
for i in range(len(com_crise)):
    raw = mne.io.read_raw_edf(path2 + com_crise.loc[i,'paths'],preload=True,verbose=False)
    sig = raw.to_data_frame().T
    sig = sig.loc[use,:].to_numpy()
    
    dur = np.array([inicio[i],fim[i]]).astype('int')
    
    for ii in range(len(dur[0])):
        mx_w = np.hstack((mx_w,sig[:,fs*dur[0,ii]:fs*dur[1,ii]]))
    
```
