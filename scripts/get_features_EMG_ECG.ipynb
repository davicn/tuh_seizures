```python
import numpy as np
import pandas as pd 
```

```python
w = np.load('w_train.npy',allow_pickle=True)
w = pd.DataFrame(w)
w.columns = ['path','start','end','type','fs','duration','montage']
print(len(w))
w.head()
```

```python
n = np.load('n_train.npy',allow_pickle=True)
n = pd.DataFrame(n)
n.columns = ['path','fs','duration','montage']
print(len(n))
n.head()
```

```python
np.unique(w['montage'])
```

```python
import mne

path = '/home/davi/TransferÃªncias/www.isip.piconepress.com/projects/tuh_eeg/downloads/tuh_eeg_seizure/v1.5.0/edf/'

fs = int(np.unique(w['fs']))
w_ecg_emg = np.zeros((2,1))


for i in range(len(w)):
    raw = mne.io.read_raw_edf(path+w.iloc[i,0].replace('tse','edf'),preload=False, verbose=False)
    d = raw.to_data_frame().T
    windows = d.loc[['EEG EKG1-REF','EMG-REF'],int(fs*.25)*int(w.iloc[i,1]):int(fs*.25)*int(w.iloc[i,2])].to_numpy()
    w_ecg_emg = np.hstack((w_ecg_emg,windows))
    print(w_ecg_emg.shape)
```

```python
n_ecg_emg = np.zeros((2,1))

for i in range(len(n)):
    raw = mne.io.read_raw_edf(path+n.iloc[i,0].replace('tse','edf'),preload=False, verbose=False)
    d = raw.to_data_frame().T
    windows = d.loc[['EEG EKG1-REF','EMG-REF'],:].to_numpy()
    n_ecg_emg = np.hstack((n_ecg_emg,windows))
    print(n_ecg_emg.shape)
    if n_ecg_emg.shape[1]>=w_ecg_emg.shape[1]:
        break
```

```python
w_ecg_emg = w_ecg_emg[:,1:]
n_ecg_emg = n_ecg_emg[:,1:]
```

```python
n_ecg_emg = n_ecg_emg[:,:w_ecg_emg.shape[1]]
print(w_ecg_emg.shape==n_ecg_emg.shape)
```

```python
np.save('w_ecg_emg.npy',w_ecg_emg)
np.save('n_ecg_emg.npy',n_ecg_emg)
```
