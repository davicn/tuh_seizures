```python
path = os.getcwd()
path2 = '/home/pib3/Projetos/tuh_eeg_seizures/'
```

### Informaçẽos de dados de treino

```python
data = pd.read_csv(path+'/docs/all_paths_test.txt',header=None).to_numpy()
data = data[:,0]
data 
```

```python
import mne 

fs = []
mont = []
time = []
for i in range(len(data)):
    raw = mne.io.read_raw_edf(path2+data[i], preload=False, verbose=False)
    fs.append(int(raw.info['sfreq']))
    mont.append(raw.info['ch_names'])
    d = raw.to_data_frame().T
    time.append(d.shape[1]//int(raw.info['sfreq']))
```

```python
info = pd.DataFrame(data=np.array([data,fs,time,mont]).T,columns=['path','freq','duracao','montage'])
info.head()
```

```python
np.save('info_test.npy',info)
```
