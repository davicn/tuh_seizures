### Features ECG

```python
from scipy.stats import skew, kurtosis

fs = int(np.unique(data['freq']))

def energy_(x):
    return np.sum(np.abs(x)**2)

def curtose(s, fs):
    m = np.array([kurtosis(s[i*fs:(i+1)*fs]) for i in range((len(s)//fs)-1)])
    return m

def assimetria(s, fs):
    m = np.array([skew(s[i*fs:(i+1)*fs]) for i in range((len(s)//fs)-1)])
    return m

def variancia(s, fs):
    m = np.array([np.var(s[i*fs:(i+1)*fs]) for i in range((len(s)//fs)-1)])
    return m

def energia(s, fs):
    m = np.array([energy_(s[i*fs:(i+1)*fs]) for i in range((len(s)//fs)-1)])
    return m
```

```python
import mne 

path2 = '/home/pib3/Projetos/tuh_eeg_seizures/dev_test/01_tcp_ar/'

ecg_features = []

for i in range(len(data)):
    raw = mne.io.read_raw_edf(path2 + data.loc[i,'paths'],preload=True,verbose=False)
    sig = raw.to_data_frame().T
    
    c = np.var(curtose(sig.loc['EEG EKG1-REF',:],fs))
    a = np.var(assimetria(sig.loc['EEG EKG1-REF',:],fs))
    v = np.var(variancia(sig.loc['EEG EKG1-REF',:],fs))
    e = np.var(energia(sig.loc['EEG EKG1-REF',:],fs))
    
    ecg_features.append(np.array([c,a,v,e]))
    
ecg_features = np.array(ecg_features)
ecg_features.shape
    

```

```python
y = data['class'].to_numpy()
y
```

```python
import seaborn as sns

x = np.hstack((ecg_features,y.reshape((len(y),1))))

X = pd.DataFrame(data=x,columns=['kurtosis','skewness','variance','energy','class'])


sns_plot = sns.pairplot(X,hue="class",size=3)
#sns_plot.savefig("resultado_preliminar.png")
X.head()
```
