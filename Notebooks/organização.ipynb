```python
import numpy as np
import pandas as pd 
import mne 
import os 
```

```python
data = np.load(path2+'data3.npy')
col = ['pacientes','edf','tse','lbl','info_seizures']
data = pd.DataFrame(data=data,columns=col)
data.head(10)
```

```python
montages = []
freq = []
duration = []
for i in range(len(data)):#len(data)):
    p1 = data.loc[i,'pacientes'][:-1]
    m1 = []
    f1 = []
    d1 = []
    for ii in range(len(data.loc[i,'edf'])):
        p2 = data.loc[i,'edf'][ii]
        p3 = path_root + p1 + p2
        raw = mne.io.read_raw_edf(p3,preload=True,verbose=False)
        m1.append(raw.info['ch_names'])
        f1.append(raw.info['sfreq'])
        d = raw.to_data_frame().shape[0]//int(raw.info['sfreq'])
        d1.append(d)
    montages.append(m1)
    freq.append(f1)
    duration.append(d1)
#         p3 = path_root + p1 + p2
        
#         raw = mne.io.read_raw_edf(p3,preload=True,verbose=False)
#         m1.append(raw.info['ch_names'])
#     montages.append(m1)
        
# p2 = data.loc[0,'edf'][0]

# p3 = path_root + p1 + p2

# raw  =mne.io.read_raw_edf(p3,preload=True,verbose=False)
# raw.info['ch_names']

print(len(montages))
print(len(freq))
print(len(duration))
```
