```python
from scipy.stats import skew, kurtosis
from numba import jit

fs = int(np.unique(data['freq']))


def energy(x):
    return np.sum(np.abs(x)**2)

@jit(nopython=True)
def mpp(x):
    return np.array([np.mean(x[:,i]) for i in range(len(x[0]))])

def curtose(s, j):
    m = np.zeros((21,len(j)-1))
    for i in range(len(m)):
        m[i] = np.array([kurtosis(s[i,j[ii]:j[ii+1]]) for ii in range(len(j)-1)])
    return m

def assimetria(s, j):
    m = np.zeros((21,len(j)-1))
    for i in range(len(m)):
        m[i] = np.array([skew(s[i,j[ii]:j[ii+1]]) for ii in range(len(j)-1)])
    return m
    
def variancia(s, j):
    m = np.zeros((21,len(j)-1))
    for i in range(len(m)):
        m[i] = np.array([np.var(s[i,j[ii]:j[ii+1]]) for ii in range(len(j)-1)])
    return m

def energia(s, j):
    m = np.zeros((21,len(j)-1))
    for i in range(len(m)):
        m[i] = np.array([energy(s[i,j[ii]:j[ii+1]]) for ii in range(len(j)-1)])
    return m

janela = lambda jfs,s:np.array([i*jfs for i in range(s//jfs+1)])
```

```python
import mne 

path2 = '/home/pib3/Projetos/tuh_eeg_seizures/dev_test/01_tcp_ar/'

eeg_features = []
for i in range(len(data)):
    raw = mne.io.read_raw_edf(path2 + data.loc[i,'paths'],preload=True,verbose=False)
    sig = raw.to_data_frame().T
    sig = sig.loc[use,:].to_numpy()

    jc = janela(fs,len(sig[0]))
    
    c = curtose(sig,jc)
    a = assimetria(sig,jc)
    v = variancia(sig,jc)
    e = energia(sig,jc)
    
    eeg_features.append(np.array([c,a,v,e]))
    
    
eeg_features = np.array(eeg_features)
```
