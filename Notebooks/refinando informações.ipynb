```python
import numpy as np
import pandas as pd 
```

```python
file = np.load('data4.npy')
col = ['pacientes','edf','tse','lbl','info_seizures','montage','freq','duration']
data = pd.DataFrame(data=file,columns=col)
data.head()
```

```python
paths = []

for i in range(len(data)):
    #print(data.loc[i,'pacientes'][:-1])
    for ii in range(len(data.loc[i,'edf'])):
        paths.append((data.loc[i,'pacientes'][:-1] + data.loc[i,'edf'][ii]))

len(paths)
```

```python
freqs = []

for i in range(len(data)):
    #print(data.loc[i,'pacientes'][:-1])
    for ii in range(len(data.loc[i,'freq'])):
        freqs.append(data.loc[i,'freq'][ii])

len(freqs)
```

```python
mont = []

for i in range(len(data)):
    #print(data.loc[i,'pacientes'][:-1])
    for ii in range(len(data.loc[i,'montage'])):
        mont.append(data.loc[i,'montage'][ii])

len(mont)
```
